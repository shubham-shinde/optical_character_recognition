```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../dataset/handwritten_math_symbols/eval/eight/21256.jpg')
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img)
img.shape, img[0][0]
```

```python
ret, img = cv2.threshold(img, 125, 255, 0)
img = 255 - img
plt.imshow(img)
```

```python
img = cv2.resize(img, (50, 50))
plt.imshow(img)
```

```python
size = np.shape(img)
skel = np.zeros(img.shape, np.uint8)
plt.imshow(skel)
```

```python
element = cv2.getStructuringElement(cv2.MORPH_CROSS, (2, 2))
print(element)
```

```python
while True:
    open = cv2.morphologyEx(img, cv2.MORPH_OPEN, element)
    temp = cv2.subtract(img, open)
    eroded = cv2.erode(img, element)
    skel = cv2.bitwise_or(skel, temp)
    plt.imshow(skel)
    plt.show()
    img = eroded.copy()
    plt.imshow(img)
    plt.show()
    print(img.shape)
    if cv2.countNonZero(img)==0:
        break
plt.imshow(skel)
    
```

```python
_, show = cv2.threshold(np.int16(skel), 127, 255, 0)
plt.imshow(show)
```
